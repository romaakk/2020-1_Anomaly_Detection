# 각각 NA infinity 제거
# 각각 error 컬럼 붙임.



```
# df.reset_index(drop=True, inplace=True)

normal1 = normal1.dropna()
normal1["error"] = 0
normal1["error"] = normal1["error"].astype(int)

normal2 = normal2.dropna()
normal2["error"] = 0
normal2["error"] = normal2["error"].astype(int)

abnormal1 = abnormal1.dropna()
abnormal1["error"] = 0
abnormal1["error"] = abnormal1["error"].astype(int)

abnormal2 = abnormal2.dropna()
abnormal2["error"] = 0
abnormal2["error"] = abnormal2["error"].astype(int)
```

# normal1, normal2, abnormal1, abnormal2를 각각 8:2로 나누고 각자 합침.
# 인덱스 리셋

```
# normal1을 8:2로 쪼갬.
x_train1 = normal1[48240:]
x_test1 = normal1[:48240]

# normal2을 8:2로 쪼갬.
x_train2 = normal2[61920:]
x_test2 = normal2[:61920]

# abnormal1을 8:2로 쪼갬.
x_train3 = abnormal1[13320:]
x_test3 = abnormal1[:13320]

# abnormal2을 8:2로 쪼갬.
x_train4 = abnormal2[58320:]
x_test4 = abnormal2[:58320]

x_train = pd.DataFrame()
x_test = pd.DataFrame()

x_train = pd.concat([x_train, x_train1])
x_train = pd.concat([x_train, x_train2])
x_train = pd.concat([x_train, x_train3])
x_train = pd.concat([x_train, x_train4])
x_test = pd.concat([x_test, x_test1])
x_test = pd.concat([x_test, x_test2])
x_test = pd.concat([x_test, x_test3])
x_test = pd.concat([x_test, x_test4])

x_train.reset_index(drop=True, inplace=True)
x_test.reset_index(drop=True, inplace=True)

x_train
```
